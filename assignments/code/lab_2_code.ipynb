```python
## QUESTION 3 ##
#parameters
start = 0
stop = 50
inc = 1
times = np.arange(start, stop+1, step=inc)
r = 1.5
K = 12.5 #enter the carrying capacity you estimated in Question 2
times
```

```python
#logistic growth function
def myLogistic(times, r, K):
    num=np.zeros(len(times)) #creates a vector of 0s that will be later replaced with actual values
    num[0]=df["Population"][0] #initial condition
    for i in range(0,len(times)-1):
        num[i+1] = num[i]+r*num[i]*(1-num[i]/K) #logistic growth expression
    return num
```

```python
num = myLogistic(times, r, K)
```

```python
plt.plot(df["Time"],df["Population"])
plt.plot(times, num)
plt.legend(["Yeast Data", "Logistic Growth with r=1.5"])
```

```python
r = 2.0
num = myLogistic(times, r, K)
plt.plot(df["Time"],df["Population"])
plt.plot(times, num)
plt.legend(["Yeast Data", "Logistic Growth with r=2.0"])
```

```python
## QUESTION 6 -- !!!PLEASE EDIT THE CODE BELOW TO REFLECT THE HARVEST!!! ##
#logistic growth + harvest function

def myHarvest(times, r, K, h):
    num=np.zeros(len(times)) #creates a vector of 0s that will be later replaced with actual values
    num[0]= K #initial condition 
    for i in range(0,len(times)-1):
        num[i+1] = num[i]+r*num[i]*(1-num[i]/K)-h #logistic growth + harvest expression
    return num
```
